```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

# Basic Graph

```python
x = [0,1,2,3,4]
y = [0,2,4,6,8]
plt.plot(x,y)
plt.title('Our first graph !')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.show()
```

```python
#what if we want to only see integer values in the X and Y axis ?

plt.plot(x,y, label ='y = 2x ', color = 'purple' , linewidth = 2 , marker = '*' , markersize = 10 , markeredgecolor = 'blue')  #you can put hexadecimal for colors  = > purple : #A020F0
plt.title('Our first graph !')
plt.xlabel('X Axis')

plt.xticks([0,1,2,3,4])
plt.yticks([0,2,4,6,8,10])

#legend
plt.legend()

plt.ylabel('Y Axis')
plt.show()
```

```python
#Different Line Styles for better visualization
plt.plot(x,y, label ='y = 2x ', color = 'purple' , linewidth = 2 , marker = '*' , markersize = 10 , markeredgecolor = 'blue' , linestyle = '--')
plt.title('Our first graph !')
plt.xlabel('X Axis')

plt.xticks([0,1,2,3,4])
plt.yticks([0,2,4,6,8,10])

#legend
plt.legend()

plt.ylabel('Y Axis')
plt.show()
```

```python
#we can use shorthand notation
# Color - marker - line

plt.plot(x,y,'b*--', label ='y = 2x')  #blue marker * line --
plt.title('Our first graph !')
plt.xlabel('X Axis')

plt.xticks([0,1,2,3,4])
plt.yticks([0,2,4,6,8,10])

#legend
plt.legend()

plt.ylabel('Y Axis')
plt.show()
```

```python
#Let's add another line to the graph

#Resize the graph

plt.figure(figsize = (5,3) , dpi = 200) #dpi changes the size of the graph
x2 = np.arange(0,4,0.5) #from zero to four with step size of 0.5

plt.plot(x2,x2**2 , 'r' , label = 'y = x^2' , marker = '^' , )
plt.plot(x,y,'b*--', label ='y = 2x')  #blue marker * line --
plt.title('Our graph !')
plt.xlabel('X Axis')

plt.xticks([0,1,2,3,4])
#plt.yticks([0,2,4,6,8,10])

#legend
plt.legend()

plt.ylabel('Y Axis')
plt.show()
```
